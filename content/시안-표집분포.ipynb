### 표집분포 위젯 (시안)

```
%pip install -q ipywidgets pandas matplotlib numpy
import pandas as pd, numpy as np, matplotlib.pyplot as plt, ipywidgets as W
pop = pd.to_numeric(pd.read_csv('data/StudentSurvey.csv')['Pulse'], errors='coerce').dropna().values
mu = pop.mean()
def f(n=10):
    m = [np.random.choice(pop, n, replace=True).mean() for _ in range(1000)]
    plt.figure(figsize=(7,4)); plt.hist(m, bins=30, color='#dc3545', alpha=.75, edgecolor='white')
    plt.axvline(mu, color='black', ls='--', lw=2); plt.xlim(mu-25, mu+25)
    plt.title(f'표본평균 표집분포 (n={n}),  SE≈{np.std(m):.2f}'); plt.show()
W.interact(f, n=W.IntSlider(min=2, max=200, value=10, description='n'));
```

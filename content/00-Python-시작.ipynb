## 위젯 예제 — 표집분포 (Sampling Distribution)

표본 크기 슬라이더를 움직이면, 표본평균의 표집분포가 어떻게 좁아지는지 실시간으로 보입니다.
(모집단 = StudentSurvey 의 `Pulse` 맥박수)

```
pop = pd.to_numeric(df['Pulse'], errors='coerce').dropna().values
mu = pop.mean()

def sampling_dist(sample_size=10, n_samples=1000):
    means = [np.random.choice(pop, sample_size, replace=True).mean() for _ in range(n_samples)]
    fig, ax = plt.subplots(figsize=(7,4))
    ax.hist(means, bins=30, color='#dc3545', alpha=.75, edgecolor='white')
    ax.axvline(mu, color='black', ls='--', lw=2, label=f'모평균 μ={mu:.1f}')
    ax.set_title(f'표본평균의 표집분포  (n={sample_size}, 반복={n_samples})')
    ax.set_xlabel('표본평균'); ax.set_ylabel('빈도')
    ax.set_xlim(mu-25, mu+25); ax.legend()
    plt.show()
    print(f'표집분포의 표준오차(SE) ≈ {np.std(means):.2f}')

W.interact(sampling_dist,
           sample_size=W.IntSlider(min=2, max=200, step=1, value=10, description='표본크기 n'),
           n_samples=W.IntSlider(min=200, max=3000, step=100, value=1000, description='반복수'));
```
